```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
from PIL import Image #converting images into arrays
from wordcloud import WordCloud, STOPWORDS
```

```python
covidwords=open('covid_words.txt','r',encoding='utf-8').read()
```

```python
stopwords=set(STOPWORDS)
```

```python
#instantiatea word cloud bject
cwordcloud=WordCloud(background_color='white',
                     max_words=3000,
                     stopwords=stopwords
                     )
#generate the word cloud
cwordcloud.generate(covidwords)
```

```python
#resize the cloud
fig=plt.figure()
fig.set_figwidth(18) #set width
fig.set_figheight(22)#set height

#display the cloud
plt.imshow(cwordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
#import virus.jpg
virus=np.array(Image.open('virus.jpg'))
```

```python
fig=plt.figure()
fig.set_figwidth(15) #set width
fig.set_figheight(18)#set height

plt.imshow(virus,cmap=plt.cm.gray, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
#instantiate a word cloud object
cwordcloud=WordCloud(background_color='white', max_words=2000,mask=virus,stopwords=stopwords)

#generate the word cloud
cwordcloud.generate(covidwords)

#display the word cloud
fig=plt.figure()
fig.set_figwidth(20)
fig.set_figheight(26)

plt.imshow(cwordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
read_file = pd.read_excel ("statewithcoordinates.xlsx")
 
# Write the dataframe object
# into csv file
read_file.to_csv ("statewithcoordinates.csv",
                  index = None,
                  header=True)
   
# read csv file and convert
# into a dataframe object
df = pd.DataFrame(pd.read_csv("statewithcoordinates.csv"))
 
# show the dataframe
df
```

```python
import folium
print ('Folium installed and imported')
latitude = 21
longitude = 78
# create map and display it
india_map = folium.Map(location=[latitude, longitude], zoom_start=4.5)

# display the map of India
india_map
```

```python
df['cases_index_quartile'] = pd.qcut(df['Confirmed'], 4, labels=False)
```

```python
colordict = {0: 'lightblue', 1: 'lightgreen', 2: 'orange', 3: 'red'}
```

```python
for lat, lon, cases_q, confirmed, state in zip(df['LATITUDE'], df['LONGITUDE'], df['cases_index_quartile'], df['Confirmed'], df['STATE']):
    folium.CircleMarker(
        [lat, lon],
        color='b',
        key_on = cases_q,
        threshold_scale=[0,1,2,3],
        fill_color=colordict[cases_q],
        fill=True,
        fill_opacity=0.7
        ).add_to(india_map)
india_map
```

```python

sns.set(font_scale=1.4)
df.set_index('total_cases')['total_death'].plot(figsize=(12, 10), linewidth=2.5, color='maroon')
plt.xlabel("total cases", labelpad=15)
plt.ylabel("total deaths", labelpad=15)
plt.title("fatality ratio", y=1.02, fontsize=22);
```

```python
sns.set(font_scale=1.4)
df.set_index('DATE')['new_cases'].plot(figsize=(15, 13), linewidth=2.5, color='maroon')
plt.xlabel("date", labelpad=20)
plt.ylabel("new cases", labelpad=20)
plt.title("new cases wrt date", y=1.02, fontsize=22);
```
