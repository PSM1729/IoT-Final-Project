```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import matplotlib.animation as animation
from IPython.display import HTML
import numpy as np
import seaborn as sns
```

```python
#Collecting the data
data = pd.read_csv("./state.csv")
```

```python
df = pd.DataFrame(data)
```

```python
#Extracting various parameters
states_ut_names = list(df['States/UT'])
total_cases = list(df['Total_cases'])
active_cases = list(df['Active_cases'])
total_deaths = list(df['Total_death'])
```

```python
plt.figure(figsize=(18, 15), dpi=80)
plt.title("COVID-19 TOTAL CASES ACROSS INDIAN STATES")
plt.barh(states_ut_names,total_cases,height=0.8,label="TOTAL CASES")

for index, value in enumerate(total_cases):
    plt.text(value, index, str(value))

plt.xlabel("Cases in numbers (10^6)")
plt.ylabel("States")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.8)
```

```python
plt.figure(figsize=(18, 15), dpi=80)
plt.barh(states_ut_names,active_cases,height=0.5,color="#D35606",label="ACTIVE CASES")

for index, value in enumerate(active_cases):
    plt.text(value, index, str(value))

plt.xlabel("Number of people")
plt.ylabel("States")
plt.title("COVID-19 ACTIVE CASES ACROSS INDIAN STATES")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.5)
```

```python
plt.figure(figsize=(18, 15), dpi=100)
plt.barh(states_ut_names,total_deaths,height=0.5,color="#6EAF46",label="TOTAL DEATHS")

for index, value in enumerate(total_deaths):
    plt.text(value, index, str(value))

plt.xlabel("Number of people")
plt.ylabel("States")
plt.title("COVID-19 TOTAL DEATHS ACROSS INDIAN STATES")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.5)
```

```python
data2 = pd.read_csv("./case_time_series.csv")
```

```python
df2 = pd.DataFrame(data2)
```

```python
#Extracting various parameters
date = list(df2['Date_YMD'])
daily_confirm = list(df2['Daily_Confirmed'])
daily_recover = list(df2['Daily_Recovered'])
daily_death = list(df2['Daily_Deceased'])
total_confirm = list(df2['Total_Confirmed'])
total_recover = list(df2['Total_Recovered'])
total_death = list(df2['Total_Deceased'])
```

```python
plt.figure(figsize=(20,10),dpi=200)

plt.plot(date,daily_confirm,color='green',linewidth = 2, label="DAILY CONFIRMED")
plt.plot(date,daily_recover,color='blue',linewidth = 2, label="DAILY RECOVERED")
plt.plot(date,daily_death,color='orange',linewidth = 2, label="DAILY DEATH")
plt.xlabel("DATES")
plt.ylabel("NO. OF PEOPLE")
plt.xticks(np.arange(0, 660, 20),fontsize=10,rotation=90)
plt.title("COVID-19 DAILY CASES ACROSS INDIAN STATES")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.8)
```

```python
plt.figure(figsize=(20,10),dpi=200)

plt.plot(date,total_confirm,color='green',linewidth = 2, label="TOTAL CONFIRMED")
plt.plot(date,total_recover,color='blue',linewidth = 2, label="TOTAL RECOVERED")
plt.plot(date,total_death,color='orange',linewidth = 2, label="TOTAL DEATH")
plt.xlabel("DATES")
plt.ylabel("NO. OF PEOPLE")
plt.xticks(np.arange(0, 660, 20),fontsize=10,rotation=90)
plt.title("COVID-19 DAILY CASES ACROSS INDIAN STATES")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.8)
```
